```python
from numpy import True_


datos_ordenmedallas=datos_medallas.sort_values(by=["Year"],ascending=False)
datos_ordenmedallas
```

```python
pie=datos_2016medallas["Team"].value_counts().plot.pie(figsize=(10,10),autopct="%0.2f %%")
pie
```
